```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path="C:/KMITL/2nd Year/2nd semester/Robot Lab 3/Assignment3/test02.pt") # Load model
model.conf = 0.5  # Set the confidence threshold for detections

class_labels = ['spoon', 'fork', 'knife'] # Define class labels

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read() # Read frame from the webcam

    results = model(frame)  # Perform object detection on the frame

    # Extract the detected objects and their labels
    objects = results.pandas().xyxy[0] # xy coordinates of the corners of the rectangle
    objects = objects[objects['name'].isin(class_labels)]
    print(objects)

    # Draw bounding boxes and labels on the frame
    for _,obj in objects.iterrows():
        x1, y1, x2, y2, conf, label = obj[['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'name']]
        x1 = int(x1); y1 = int(y1); x2 = int(x2); y2 = int(y2)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f'{label}: {conf:.2f}', (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with bounding boxes and labels
    cv2.imshow('Object Detection', frame)

    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
